### 1. Perceptr贸n

```python
# Importamos la(s) libreria(s) necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from sklearn.preprocessing import StandardScaler
```

a) Escriba una funci贸n que calcule el valor de salida (output) del modelo f(x) para un patr贸n de entrada  a trav茅s de los pesos  del modelo. Decida si incluir los bias dentro de  o manejarlos de manera separada.

```python
# Creamos la funcion del ouput donde consideramos los bias dentro de w
def f(X,W):
    if np.dot(X,W) >=0: # Producto punto entre X y W
        return 1
    else:
        return 0
```

b) Escriba una funci贸n que implemente el cl谩sico algoritmo del Perceptr贸n para un problema binario que permita entrenarlo en un conjunto de datos de tama帽o N, le铆dos de manera online (uno a uno). Recordar la decisi贸n anterior sobre los bias.

```python
def perceptron(X, y, W, etha):
    if y != f(X,W):
        if y == 1:
            W = W + etha*X
        else:
            W = W - etha*X
    return W
```

c) Demuestre que lo implementado funciona en un problema real de clasificaci贸n. Para esto utilice el dataset Breast cancer wisconsin, disponible a trav茅s de la librer铆a sklearn

```python
# Importar dataset
from sklearn.datasets import load_breast_cancer
X_train,y_train = load_breast_cancer(return_X_y=True)

#normalizar X_train
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
#agregar columna de 1's
N,M = X_train.shape
X_train = np.c_[X_train, np.ones(N)]

#crear clasificador
W = np.zeros(M+1)       # random initial guess
accuracy = np.zeros(N)# accuracy per iteration
#entrenar al clasificador
for i in range(N):
    W = perceptron(X_train[i],y_train[i],W,1)
    aciertos = 0
    #calcular accuracy para cada step
    for j in range(N):
        if y_train[j] == f(X_train[j],W):
            aciertos += 1
    accuracy[i] = aciertos/N
    
#graficar accuracy vs n煤mero de iteraciones
plt.figure(figsize=(14,7))
plt.plot(range(N),accuracy)
plt.xlabel("Iteration")
plt.ylabel("Accuracy")
plt.show()
```

d) Escriba una funci贸n que implemente el **Forgetr贸n** con una memoria de tama帽o $K$ y la funci贸n de kernel como el producto interno (*inner-product*), esto es $<a,b> = \sum_i a_i \cdot b_i$.

```python
#Forgetron
def forgetron(X,y, etha,K):
    I = dict()
    W = dict()
    Q = 0
    M = 0
    n,m = X.shape
    w = np.zeros(m)
    # accuracy per iteration
    accuracy = np.zeros(N)      
    for t in range(n):
        # si el modelo se equivoca
        if y[t] != f(X[t],w):
            if y[t] == 1:
                u = 1
            else:
                u = -1
            M = M + 1
            #hacer update
            W[t] = w + etha*u*X[t]
            if len(I) < K:
                I[t] = 1    
            #si no se tiene memoria
            else:
                #calcular nuevos pesos
                r = min(I, key=I.get)
                #test = np.roots([I[r]**2-2*u*I[r],2*I[r],Q-(15/32)*M])
                #test = np.append(test,0)
                #p = np.amax(test)
                #p = min(1,p)
                #p = max(0,p)
                
                #lo intent茅
                PHI = lambda p: (I[r]*p)**2 + 2*I[r]*p*(1-p*u)
                p = 1
                while p > 0:
                    if PHI(p) + Q < (15/32)*M:
                        break
                    p = p-0.001
                    
                I[t] = 1
                for i in I:
                    I[i] = p*I[i]
                
                Q = Q + PHI(p)
                #eliminar el peso menos relevante
                del I[r]
                del W[r]
        #recalcular w
        w = np.zeros(m)
        for i in W:
            w = w + W[i]*I[i]
        #calcular accuracy para cada step
        aciertos = 0
        for j in range(N):
            if y[j] == f(X[j],w):
                aciertos += 1
        accuracy[t] = aciertos/N
    return w,accuracy
```

f) Escriba una funci贸n que compute la funci贸n sigmoidal para una entrada $\xi$ cualquiera. *Tenga cuidado con los l铆mites de n煤meros que puede trabajar python (por ejemplo $\exp{800}\rightarrow +\infty$)*. *Se aconseja acotar/truncar los valores que entran a la funci贸n para que la operaci贸n se pueda realizar*. Adem谩s escriba una funci贸n que calcule la salida del nuevo modelo $g(x; w,b)$ con esta funci贸n sigmoidal.

```python
# Creamos la funcion sigmoidal
def sigmoidal(z):
    if z>=100:
        return 1
    if z<=-100:
        return 0
    return 1/(1+np.exp(-z))

# Observamos el comportamiento de la funcion sigmoidal
res = 100
_X = np.linspace(-50,50,res)
_Y = np.zeros(res)

for ix,x in enumerate(_X):
    _Y[ix]=sigmoidal(x)

plt.plot(_X,_Y),plt.grid(),plt.title("Comportamiento de la funci贸n sigmoidal")
plt.text(-45,0.6,r"$f(x)=\frac{1}{1+e^{-\beta x}}$",fontsize=20)
plt.show()

# Creamos la funcion del ouput donde consideramos los bias dentro de w
def f(X,W):
    return sigmoidal(np.dot(X,W))
```

g) Escriba una funci贸n que calcule la funci贸n de p茅rdida descrita anteriormente para un dato $x^{(i)}$, utilizando $g(x^{(i)};w,b)$. *Tenga cuidado con los l铆mites del logaritmo (recordad que $\log{0}\rightarrow +\infty$)*.

```python
# Creamos la funcion de perdida
def loss(y,X,W):
    if y==1:
        return -np.log(f(X,W))  
    else:
        return -np.log(1-f(X,W))

# Observamos el comportamiento de la funcion loss
def lossview(y,x):
    if y==1:
        return -np.log(x)  
    else:
        return -np.log(1-x)
    
res = 100
_X = np.linspace(0.01,0.99,res)
_E1 = np.linspace(1,1,res)
_E2 = np.linspace(0,0,res)
_Y1 = np.zeros(res)
_Y2 = np.zeros(res)

for ie,e in enumerate(_E1):
    _Y1[ie]=lossview(e,_X[ie])

for ie,e in enumerate(_E2):
    _Y2[ie]=lossview(e,_X[ie])    
    
plt.subplot(221)
plt.plot(_X,_Y1),plt.grid()
plt.title("Comportamiento de la funci贸n loss cuando la entrada es 1")
plt.show()

plt.subplot(222)
plt.plot(_X,_Y2),plt.grid()
plt.title("Comportamiento de la funci贸n loss cuando la entrada es 0")
plt.show()
```

h) Escriba una funci贸n que calcule el gradiente (derivada) de la funci贸n de p茅rdida anterior con respecto a los pesos del modelo $w$. *Se recomienda derivarla anal铆ticamente y luego escribirla*. *Recuerde el uso de la regla de la cadena*.

```python
def deriv(y,X,W):
    return (f(X,W)-y)*X
```

i) Realice una modificaci贸n al algoritmo implementado en b) (**Perceptr贸n**) para que se adapte a la funci贸n objetivo *binary cross entropy* implementada, para 茅sto haga uso del algoritmo de optimizaci贸n SGD [[4]](#refs) (*Stochastic Gradient Descend*) con tasa de aprendizaje $\eta \in [0,1]$.

```python
def perceptron(X,W,y,lr):
    grad = deriv(y,X,W)
    W = W - lr*grad
    return W
```
